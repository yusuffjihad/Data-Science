```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(X):
    # Caulating VIF
    vif = pd.DataFrame()
    vif["variable"] = X.columns
    vif["VIF"] = [variance_inflation_factor (X.values, i) for i in range(X.shape [1])]
    return (vif)
     
```

```python
calc_vif(dataset[["etest_p","salary"]])
```

### T-Test

## Independent Sample - Unpaierd T_Test
Different Group (Male) but same condition ('salary')

```python
from scipy.stats import ttest_ind
dataset=dataset.dropna()
male=dataset[dataset['gender']=='M']['salary']
female = dataset[dataset['gender']=='F']['salary']
#print (male)
ttest_ind(male,female)
```

## Dependent Sample - Paierd T_Test
Same Group (Male) but Differnernt condition ('ssc_p','hsc_p')

```python
from scipy.stats import ttest_rel
#dataset=dataset.dropna()
male=dataset[dataset['gender']=='M']['ssc_p']
male = dataset[dataset['gender']=='F']['hsc_p']
ttest_rel(male, male)
```

 ###Independent Sample - Unpaierd T_Test

Different Group (Commerce) but same condition ('etest_p')
Different Group (Science) but same condition ('etest_p')

```python
from scipy.stats import ttest_ind
#dataset=dataset.dropna()
Commerce=dataset[dataset['hsc_s']=='Commerce']['etest_p']
Science = dataset[dataset['hsc_s']=='Science']['etest_p']
ttest_ind(Commerce,Science)
```

## Dependent Sample - Paierd T_Test
Same Group (Male) but Differnernt condition ('ssc_p','hsc_p')

```python
from scipy.stats import ttest_rel
#dataset=dataset.dropna()
Commerce=dataset[dataset['hsc_s']=='Commerce']['etest_p']
Commerce1=dataset[dataset['hsc_s']=='Commerce']['etest_p']
ttest_rel(Commerce,Commerce1)
```
